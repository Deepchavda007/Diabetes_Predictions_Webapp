# Importing Modules

```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

# Data Collection

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

# Split dataset into training set and test set

```python
x = data.iloc[:,:-1]
y = data['Outcome']
x
```

```python
# Dividing input and output data to train and test data
# Training : Testing = 80 : 20
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, 
                                                     random_state = 42)
xtrain
```

# Fit LogisticRegression Model

```python
model = make_pipeline(StandardScaler(), LogisticRegression())
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict(xtest)
ypred
```

# Measure Accuracy

```python
testing_data_model_score = model.score(xtest, ytest)
print("Model Score/Accuracy on Testing data",testing_data_model_score)

training_data_model_score = model.score(xtrain, ytrain)
print("Model Score/Accuracy on Training data",training_data_model_score)
```

```python
accuracyscore = accuracy_score(ypred, ytest)

print("Accuracy score of model is :" ,accuracyscore*100)
```

```python
classificationreport = classification_report(ypred, ytest)

print("classification Report of model is :" ,classificationreport)
```

# Building a Predictive System

```python
input = np.array([6,148,72,35,0,33.6,0.627,50,])
input_reshape = input.reshape(1, -1)
pred = model.predict(input_reshape)

print(pred)

if (pred[0]== 0):
  print('The Person does not have a Diabetas')
else:
  print('The Person has Diabetas')
```
